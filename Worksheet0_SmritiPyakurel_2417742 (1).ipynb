## Task 1: Unit Converter Program
This program converts units for length, weight and volume using user input.

```python
# Unit Converter

def unit_converter():
    try:
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Choose conversion type: ")
        value = float(input("Enter value: "))

        if choice == "1":
            print("1. Meter to Feet")
            print("2. Feet to Meter")
            op = input("Choose option: ")

            if op == "1":
                print("Result:", value * 3.28)

            elif op == "2":
                print("Result:", value / 3.28)

        elif choice == "2":
            print("1. Kg to Pound")
            print("2. Pound to Kg")
            op = input("Choose option: ")

            if op == "1":
                print("Result:", value * 2.2)

            elif op == "2":
                print("Result:", value / 2.2)

        elif choice == "3":
            print("1. Liter to Gallon")
            print("2. Gallon to Liter")
            op = input("Choose option: ")

            if op == "1":
                print("Result:", value * 0.26)

            elif op == "2":
                print("Result:", value / 0.26)

        else:
            print("Invalid choice")

    except:
        print("Invalid input")

unit_converter()
```

## Task 2: Mathematical Operations on List
This program calculates sum, average, maximum and minimum values from a list of numbers.

```python
def find_sum(lst):
    return sum(lst)

def find_avg(lst):
    return sum(lst)/len(lst)

def find_max(lst):
    return max(lst)

def find_min(lst):
    return min(lst)

nums = list(map(int, input("Enter numbers: ").split()))

print("1. Sum")
print("2. Average")
print("3. Max")
print("4. Min")

choice = input("Choose: ")

if choice == "1":
    print(find_sum(nums))

elif choice == "2":
    print(find_avg(nums))

elif choice == "3":
    print(find_max(nums))

elif choice == "4":
    print(find_min(nums))
```

#List Manipulation
1. Extract every other element

```python
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other([1,2,3,4,5,6]))
```

2. Get sublist

```python
def get_sublist(lst,start,end):
    return lst[start:end+1]

print(get_sublist([1,2,3,4,5,6],2,4))
```

3. Reverse list

```python
def reverse_list(lst):
    return lst[::-1]

print(reverse_list([1,2,3,4,5]))
```

4. Remove first and last

```python
def remove_first_last(lst):
    return lst[1:-1]

print(remove_first_last([1,2,3,4,5]))
```

5. First n elements

```python
def get_first_n(lst,n):
    return lst[:n]

print(get_first_n([1,2,3,4,5],3))
```

6. Last 6 elements

```python
def get_last_n(lst,n):
    return lst[-n:]

print(get_last_n([1,2,3,4,5],2))
```

7. Reverse skip

```python
def reverse_skip(lst):
    return lst[-2::-2]

print(reverse_skip([1,2,3,4,5,6]))
```

4.3 Exercise on Nested List
1 Flatten Nested List

```python
def flatten(lst):
    result = []
    for sub in lst:
        for item in sub:
            result.append(item)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

2 Access Nested Element

```python
def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

3 Sum of Nested List

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item,list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

4 Remove Element from Nested List

```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item,list):
            result.append(remove_element(item,elem))
        elif item != elem:
            result.append(item)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

5 Find Maximum Element

```python
def find_max(lst):
    maximum = float('-inf')

    for item in lst:
        if isinstance(item,list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum,item)

    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

6. Count Occurrences

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item,list):
            count += count_occurrences(item,elem)
        elif item == elem:
            count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

7 Deep Flatten

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item,list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

8 Nested List Average

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

Task 1: Array Creation

```python
# empty 2x2
a = np.empty((2,2))

# ones 4x2
b = np.ones((4,2))

# full array
c = np.full((3,3),7)

# zeros like
d = np.zeros_like(c)

# ones like
e = np.ones_like(c)

# list to numpy array
new_list = [1,2,3,4]
f = np.array(new_list)

print(a,b,c,d,e,f)
```

Task 2: Array Manipulation

```python
# 10 to 49
a = np.arange(10,50)

# 3x3 matrix
b = np.arange(9).reshape(3,3)

# identity
c = np.eye(3)

# random array mean
d = np.random.random(30)
print(d.mean())

# 10x10 random min max
e = np.random.random((10,10))
print(e.min(),e.max())

# zero array replace
f = np.zeros(10)
f[4] = 1

# reverse array
arr = np.array([1,2,0,0,4,0])
print(arr[::-1])

# border array
g = np.ones((5,5))
g[1:-1,1:-1] = 0

# checkerboard
h = np.zeros((8,8))
h[1::2,::2] = 1
h[::2,1::2] = 1

print(a,b,c,f,g,h)
```

Task 3: Array Operations

```python
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11,12])

# addition
print(x + y)

# subtraction
print(x - y)

# multiply with integer
print(x * 2)

# square
print(x**2)

# dot products
print(np.dot(v,w))
print(np.dot(x,v))
print(np.dot(x,y))

# concatenate
print(np.vstack((x,y)))
print(np.column_stack((v,w)))
```

Task 4: Matrix Operations

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# A * A^-1
A_inv = np.linalg.inv(A)
print(np.dot(A,A_inv))

# AB != BA
print(np.dot(A,B))
print(np.dot(B,A))

# (AB)^T = B^T A^T
print(np.transpose(np.dot(A,B)))
print(np.dot(B.T,A.T))
```

Performance Experiment (NumPy vs Python)

```python
import time
import numpy as np

size = 1000000

a = list(range(size))
b = list(range(size))

# Python list addition
start = time.time()
c = [a[i] + b[i] for i in range(size)]
print("List addition time:", time.time()-start)

# numpy addition
a_np = np.array(a)
b_np = np.array(b)

start = time.time()
c = a_np + b_np
print("Numpy addition time:", time.time()-start)
```
